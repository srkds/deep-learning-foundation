```python
def der_of_f(x):
  return 2*(x**(2-1))
```

```python
x = [1,1.5,1.9,2,3,4.5]

y = [f(i) for i in x]
```

```python
# Define tangent line
# y = m*(x - x1) + y1
def line(x, x1, y1):
    return der_of_f(x1)*(x - x1) + y1
```

```python
def hunger(time):
  return (time**2)+(1/2)


def der_hunger(time):
  return 2*time
```

```python
time = [1,1.90,2,2.40,3,3.80,4.20,4.90,5.10,5.5,6]
hunger = [hunger(i) for i in time]
```

```python
plt.plot(time, hunger)
```

```python
def craves_icec(hunger):
  return math.sqrt(hunger)

def der_of_craves(time):
  return 1/2 * math.sqrt(hunger(time))
```

```python
cfic = [craves_icec(i) for i in hunger]

plt.plot(hunger, cfic)
```

```python
der_craves_over_time = der_of_craves(40) * der_hunger(40)
der_craves_over_time
```

# Example 3

```python
def height(weight, intercept):
  return intercept + (1*weight)
```

```python
actual_height = 2
intercept = [0,0.1,0.3,0.5,0.6,0.7,0.9,0.99,1,1.4,1.9,2]
weight = 1

predicted_vals = [height(weight,i) for i in intercept]
xrange = np.linspace(weight-1, weight+1, 10)
yrange = [height(i,intercept) for i in xrange]
```

```python
# y = height(weight,intercept)
plt.scatter([weight]*12,predicted_vals)
plt.plot(xrange, yrange)

plt.scatter([weight],[actual_height], color="red")
```

```python
# residule = actual_height - y
residule = [actual_height - pv for pv in predicted_vals]
residule_squre = [r**2 for r in residule]
print(residule)
print(residule_squre)
```

```python
plt.scatter([intercept],[residule])
plt.xlabel("intercept")
plt.ylabel("residule")
```

```python
plt.scatter([residule],[residule_squre])

plt.xlabel("residule")
plt.ylabel("residule_squre")
```
